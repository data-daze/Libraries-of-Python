```python
import matplotlib.pyplot as plt

x=["pyhton","c","c++","java"]
y=[85,60,65,75]

plt.xlabel("languages",fontsize=20)
plt.ylabel("number",fontsize=20)
plt.title("bar plot",fontsize=20)

c=["orange","brown","olive","m"]
plt.bar(x,y, width=0.4,color=c,align="edge",edgecolor='blue',linewidth=3,linestyle=":",alpha=0.7,label="tama")#or align='center'# edgecolor means boundary
plt.legend()
#plt.bar(x,y, width=0.4,color='orange')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["pyhton","c","c++","java"]
y=[85,60,65,75]
z=[20,30,40,50]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=20)
plt.ylabel("number",fontsize=20)
plt.title("bar plot",fontsize=20)

c=["orange","brown","olive","m"]
plt.bar(p,y, width, color=c,label="popularity")
plt.bar(p1,z, width, color="y",label="popularity1")



plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
#scatter plot

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]

y=[1,2,3,4,5,4,7]
colors=[20,30,40,60,70,50,80]
sizes=[470,660,1280,450,600,1000,890]
plt.scatter(x,y,  c=colors,s=sizes, marker="*", edgecolor="black", linewidth=2, cmap="viridis")#alpha=0.6
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter plot", fontsize=15)
plt.xlabel("Day" , fontsize=15)
plt.ylabel("Task no" , fontsize=15)

plt.show()
```

```python
# histrogram port


import matplotlib.pyplot as plt
import numpy as np
import random

x=np.random.randint(20,60,(25))
print(x)


no=[23 ,23,58, 49, 30, 49, 56, 33, 33, 21, 46, 52, 56, 22, 42, 24, 52, 24, 23, 44, 56, 54, 50, 50, 39,
 47]
l=[20,30,40,50,60]

plt.hist(no,edgecolor="black", linewidth=2, bins=l,cumulative=-1, bottom=20,align="mid",rwidth=0.8,log=True,label="python")
plt.title("histrogram")
plt.xlabel("python")
plt.ylabel("no")
plt.axvline(45,color="r")

plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
area1=[2,2,4,4,5,5]
area2=[1,2,3,3,4,5]
area3=[1,4,2,6,3,5]
l=["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3, colors=["y","g","orange"], labels=l)
plt.legend(loc=2)
plt.xlabel("x-axis")
plt.ylabel("area")
plt.title("python")
plt.show()
```

```python
#Fill_bewteen plot

import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y, color="black")

plt.fill_between(x,y, color="g" ,alpha=0.5)#where=(x>=2) & (x<=4),

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
