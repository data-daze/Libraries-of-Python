```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 10)
y = np.sin(x)
y2 = np.random.rand(10)
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]
distribution = np.random.randn(1000)

# Light color palette
colors = ['#A2C5AC', '#F6D6AD', '#FFBCBC', '#A0CED9', '#FFE156']

fig, axs = plt.subplots(5, 2, figsize=(14, 20))
fig.suptitle('Matplotlib Basic Plot Types - Light Professional Theme', fontsize=16)

# Line Plot
axs[0, 0].plot(x, y, color=colors[0], marker='o')
axs[0, 0].set_title('Line Plot')

# Scatter Plot
axs[0, 1].scatter(x, y2, color=colors[1])
axs[0, 1].set_title('Scatter Plot')

# Bar Chart
axs[1, 0].bar(categories, values, color=colors)
axs[1, 0].set_title('Vertical Bar Chart')

# Horizontal Bar Chart
axs[1, 1].barh(categories, values, color=colors)
axs[1, 1].set_title('Horizontal Bar Chart')

# Histogram
axs[2, 0].hist(distribution, bins=20, color=colors[2], edgecolor='black')
axs[2, 0].set_title('Histogram')

# Box Plot
axs[2, 1].boxplot([distribution[:250], distribution[250:500], distribution[500:]], patch_artist=True,
                  boxprops=dict(facecolor=colors[3]))
axs[2, 1].set_title('Box Plot')

# Violin Plot
axs[3, 0].violinplot([distribution[:250], distribution[250:500], distribution[500:]])
axs[3, 0].set_title('Violin Plot')

# Stem Plot
markerline, stemlines, baseline = axs[3, 1].stem(x, y, basefmt=" ")
plt.setp(markerline, marker='o', color=colors[4])
plt.setp(stemlines, color=colors[4])
axs[3, 1].set_title('Stem Plot')

# Step Plot
axs[4, 0].step(x, y, color=colors[0])
axs[4, 0].set_title('Step Plot')

# Pie Chart
axs[4, 1].pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140)
axs[4, 1].set_title('Pie Chart')

plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

Matplotlib is a widely used Python library for data visualization. It provides a variety of functions for creating static, animated, and interactive plots. It is especially useful for:

* Creating line plots, bar charts, histograms, scatter plots, pie charts, and more.

* Customizing plots with titles, labels, legends, colors, styles, and annotations.

* Saving plots in various formats (e.g., PNG, PDF, SVG).


**When is Matplotlib used?**

You typically use Matplotlib when:

* You want to explore datasets visually to identify patterns, trends, or outliers.
* You need to present results in reports or publications with visual support.

* You're working with other libraries like NumPy, Pandas, or SciPy and need flexible visualization tools.












**Basic Plot Types**

barh() – Horizontal bar chart

hist() – Histogram

boxplot() – Box and whisker plot

violinplot() – Violin plot

stem() – Stem plot

step() – Step plot

pie() – Pie chart







# **plot() – Line Plot**


What is plot() – Line Plot?

The plot() function in Matplotlib is used to create line plots, which connect a series of data points with straight lines. It is ideal for visualizing continuous data, trends, and relationships between variables over time or sequence.

```python
# Grid coordinates
x, y = np.meshgrid(np.arange(0, 3), np.arange(0, 3))

# Vector components (u, v)
u = np.array([[1, 0, -1],
              [0, 1, 0],
              [-1, 0, 1]])
v = np.array([[0, 1, 0],
              [-1, 0, 1],
              [0, -1, 0]])

plt.figure(figsize=(6, 6))
plt.quiver(x, y, u, v, color='#FF7F50', scale=3, width=0.005, alpha=0.8)
plt.title('Wind Velocity Vector Field')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python


# Grid coordinates
Y, X = np.mgrid[-3:3:100j, -3:3:100j]

# Vector field components (circular flow)
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.figure(figsize=(7, 7))
# Pass alpha within the color tuple
plt.streamplot(X, Y, U, V, color=(100/255, 149/255, 237/255, 0.8), density=1.5, linewidth=1.5, arrowsize=1)
plt.title('Fluid Flow Streamlines')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
