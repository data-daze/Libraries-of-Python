```python
# Line plot

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var1=[1,2,3,4,5,6]
var2=[2,3,4,4,1,6]

x=pd.DataFrame({"var1":var1,"var2":var2})
sns.lineplot(x="var1",y="var2",data=x)
plt.show()
```

# pair plot

```python
#During the eritting mechinel learning algorithm, it is important  to compaire all ghraphes.There it is used.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

var=sns.load_dataset("tips")
var
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="BuGn")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"],kind="reg",diag_kind="hist")#kde,scatter,hist,reg,
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],markers=["*","<"])
plt.show()
```

# Box plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time",order=["Fri","Sun","Sat","Thur"],showmeans=True,
           meanprops={"marker":"*","markeredgecolor":"y"},palette="plasma",linewidth=2,orient="v")#color="r"
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

var=sns.load_dataset("tips").head(20)
var
```
