# One Hot Encoding

What is encoding?

If we have any categorical data and we  connvert it as a numerical data,then it is known encoding .
We use encoding technique because we want to use data in ml algorithm.And ML algorithm is a mathmatical formula only.

# For calculating on mathematical formula we need numerical data.If we gives categorial data, then it won't work.For this condition, we use encoding concept to convert categorical data to numerical data.

One hot encoding is be done if number of data is less.For instance, in Gender column there may two types of data only like Male and Female.

# Ordinal Encoding

There are two precess to ordinal encoding.First, sklearn.Second one is Map Function.

```python
#Ordinal Encoding by sklearn

df=pd.DataFrame({"Size":["s","l","m","xl","m"]})
df
```

```python
oe.transform(df[["Size"]])
```

# Outlier remover using IQR

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
ds=pd.read_csv("loan.csv")
ds.head(3)

```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x="CoapplicantIncome",data=ds)
plt.show()
```

```python
ds.shape
```

```python
q1=ds["CoapplicantIncome"].quantile(0.25)
q3=ds["CoapplicantIncome"].quantile(0.75)

IQR=q3-q1

```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)

min_range,max_range
```

```python
ds[ds["CoapplicantIncome"]<=max_range] # here we can get those data which value are less then the max_range
```

```python
new_ds=ds[ds["CoapplicantIncome"]<=max_range] # this is the main line to remove outliers
```

```python
new_ds.shape
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_ds)
plt.show()
```

Sometimes, when we remove outlier, maybe it will remove some important features.So it ie quite important to remove less outliers.

```python
min_range,max_range
```

# Feature Scaling Techniques

Suppose , we have a dataset.Which has values 1000,2000,30000, and 2.50,0.9,6.0 etc.when we will give this dataset into a ml model, it will work with larger data(1000,2000 etc).
But both data are imporatnt here.For remove this problem, we use feature scaling techniques.There are two types of feature scaling.First, standardizaton.Second,Normalization

standardizaton and Normalization work differently

# standardizaton.

```python
ds=pd.read_csv("loan.csv")
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
ds["ApplicantIncome"].fillna(ds["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.distplot(ds["ApplicantIncome"])
plt.show()
```

```python
ds.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(ds[["ApplicantIncome"]])
```

```python
ss.transform(ds[["ApplicantIncome"]])
```

```python
ds["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(ds[["ApplicantIncome"]]),columns=["x"])
```

```python
ds.head(3)
```

```python
ds.describe()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.distplot(ds["ApplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(ds["ApplicantIncome_ss"])
plt.title("After")

plt.show()
```

After scaling, the magnitude of data will be changed.But the nature of the data will be same as before.
